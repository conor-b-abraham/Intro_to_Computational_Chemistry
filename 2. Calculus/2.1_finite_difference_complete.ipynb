# 3-point finite difference method #

In this exercise we will create a function to approximate the first derivative of $cos(x)$ using the 3-point finite difference formula:

$$ f'(x) = \frac{f(x+h)-f(x-h)}{2h} $$

Then, we will compare that to the exact value. 

```python
import numpy as np
def dcosx_dx(x, delx):
    '''
    Approximates the first derivative of cos(x) using the 3-point finite difference formula. 

    f'(x) = [f(x+h)-f(x-h)]/(2h)

    Parameters
    ----------
    x : float or numpy.ndarray(dtype=float)
        The point of differentiation. If numpy.ndarray and delx is numpy.ndarray, must be same shape as delx.
    delx: float or numpy.ndarray(dtype=float)
        The finite difference element. If numpy.ndarray and x is numpy.ndarray, must be same shape as x.

    Returns
    -------
    numerical : float or numpy.ndarray(dtype=float)
        The approximated value using the 3-point finite difference formula. Is a float if both h and x are floats, otherwise is a numpy array.
    exact : float or numpy.ndarray(dtype=float)
        The exact value of the derivative. = -sin(x). Is a float if x is a float, is a numpy array if x is a numpy array.
    '''
    numerical = (np.cos(x+delx)-np.cos(x-delx))/(2*delx) 
    exact = -np.sin(x)

    return numerical, exact
```

Create a plot comparing the exact values of $\frac{d}{dx}cos(x)$ to the approximated values obtained from the 3-point formula for x from 0 to 2$\pi$:

```python
x_vals = np.linspace(0, 2*np.pi, 180)
numerical, exact = dcosx_dx(x_vals, 0.5)

import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x_vals, numerical, label="3-point")
ax.plot(x_vals, exact, label="Exact")

ax.set_xticks([0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
               [0, r"$\frac{\pi}{2}$", r"$\pi$", r"$\frac{3\pi}{2}$", r"$2\pi$"])

ax.set_xlabel("x")
ax.set_ylabel("cos(x)")

ax.legend()

plt.show()
```

Now, create a plot showing the error in the 3-point difference formula for x = 1 and $\Delta x$ ranging from $10^{-5}$ to $\frac{1}{2}$, where:

$$ error = \frac{d}{dx}cos(x)\; \mathrm{(exact)} - \frac{d}{dx}cos(x) \; \mathrm{(numerical)} $$

```python
delx_vals = np.linspace(10**-5, 1/2, 100)
numerical, exact = dcosx_dx(1, delx_vals)

fig, ax = plt.subplots()
ax.plot(delx_vals, -np.abs(exact-numerical)/exact, label="3-point")
ax.set_xlabel("h")
ax.set_ylabel("Fractional Error")

ax.set_xlim(10**-5, 0.5)

plt.show()
```
